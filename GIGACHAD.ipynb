```python
URL = 'https://query1.finance.yahoo.com/v7/finance/download/BTC-USD?period1=1601916783&period2=1633452783&interval=1d&events=history&includeAdjustedClose=true'
```

```python
from urllib.request import urlretrieve
urlretrieve(URL,"bit.csv")
```

```python
!head bit.csv
```

```python
import pandas as pd
df=pd.read_csv("bit.csv",index_col="Date",parse_dates=True)
```

```python
df.shape
```

```python
df.resample("W").sum().plot()
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
df.resample("W").sum().plot()
```

```python
ax=df.resample('D').sum().rolling(10).sum().plot()
ax.set_ylim(0,None)
```

```python
df.groupby(df.index).mean().plot()
```
